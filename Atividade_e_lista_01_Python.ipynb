```python
#13. Elabore um algoritmo que leia um número de entrada que indicará a quantidade de registros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de N pessoas e ao final apresentar a média de idade de ambos os gêneros catalogados.

quantidade_registros = int(input("Quantas pessoas você irá registrar? "))

soma_idade_m = 0
contador_m = 0
soma_idade_f = 0
contador_f = 0
contador_registros = 1

if quantidade_registros > 0:
    while contador_registros <= quantidade_registros:
        print(f"\n--- Inserindo dados da {contador_registros}ª pessoa ---")

        sexo = input("Digite o sexo (M/F): ").upper()
        idade = int(input("Digite a idade: "))

        if sexo == 'M':
            soma_idade_m += idade
            contador_m += 1
        elif sexo == 'F':
            soma_idade_f += idade
            contador_f += 1
        else:
            print("Sexo inválido. Este registro será ignorado.")

        contador_registros += 1

    if contador_m > 0:
        media_idade_m = soma_idade_m / contador_m
        print(f"Média de idade dos homens: {media_idade_m:.2f} anos.")
    else:
        print("Nenhum homem foi registrado.")

    if contador_f > 0:
        media_idade_f = soma_idade_f / contador_f
        print(f"Média de idade das mulheres: {media_idade_f:.2f} anos.")
    else:
        print("Nenhuma mulher foi registrada.")

else:
    print("Nenhum registro foi solicitado.")
```

```python
#15. Elabore um algoritmo que solicite N números reais e quando o usuário informar o valor nulo 0 (zero) o programa ordene e mostre todos os números informados de forma crescente.

numeros = []

print("Digite quantos números reais desejar.")
print("Para finalizar e ordenar, digite o número 0 (zero).")

while True:
    numero_lido = float(input("Digite um número: "))

    if numero_lido == 0:
        break

    numeros.append(numero_lido)

if numeros:
    numeros.sort()

    print(*numeros, sep='\n')
else:
    print("\nNenhum número foi informado para ordenação.")
```

```python
#16. Escreva um programa que vá solicitando as idades dos alunos da sala até que todos sejam informados (perguntar ao usuário se deseja informar a idade do próximo aluno). Ao final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de 18 anos, quantos alunos têm até 18 anos, a média aritmética e a mediana.

idades = []

print("--- Cadastro de Idades dos Alunos ---")

while True:
    idade_str = input("\nDigite a idade do aluno: ")

    if idade_str.isdigit():
        idade_atual = int(idade_str)
        idades.append(idade_atual)
    else:
        print("Entrada inválida. Por favor, digite um número inteiro positivo.")
        continue

    while True:
        continuar = input("Deseja informar a idade do próximo aluno? (S/N): ").upper()
        if continuar == 'S' or continuar == 'N':
            break
        else:
            print("Opção inválida. Por favor, digite S para Sim ou N para Não.")

    if continuar == 'N':
        break

if idades:

    idades.sort()

    mais_novo = idades[0]
    mais_velho = idades[-1]

    maiores_de_18 = 0
    ate_18 = 0
    indice = 0
    total_alunos = len(idades)

    while indice < total_alunos:
        idade = idades[indice]
        if idade > 18:
            maiores_de_18 += 1
        else:
            ate_18 += 1
        indice += 1

    soma_idades = sum(idades)
    media = soma_idades / total_alunos

    # Calculate the median
    meio = total_alunos // 2
    if total_alunos % 2 == 0:
        mediana = (idades[meio - 1] + idades[meio]) / 2
    else:
        mediana = idades[meio]


    print(f"A idade do aluno mais novo é: {mais_novo} anos")
    print(f"A idade do aluno mais velho é: {mais_velho} anos")
    print(f"Quantidade de alunos com mais de 18 anos: {maiores_de_18}")
    print(f"Quantidade de alunos com até 18 anos: {ate_18}")
    print(f"Média de idade da turma: {media:.2f} anos")
    print(f"Mediana das idades: {mediana:.2f} anos")

else:
    print("\nNenhuma idade foi informada para gerar estatísticas.")
```
